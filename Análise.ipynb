##Período do dia

```python
def periodo_do_dia(h):
    if 0 <= h < 6:
        return 'Madrugada'
    elif 6 <= h < 12:
        return 'Manhã'
    elif 12 <= h < 18:
        return 'Tarde'
    else:
        return 'Noite'
```

```python
data_filtrada['periodo'] = data_filtrada['hora_abertura'].apply(periodo_do_dia)
data_filtrada['periodo'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Chamados por Período do Dia')
plt.xlabel('Período')
plt.ylabel('Total de Chamados')
plt.grid(axis='y')
plt.show()
```

# Organização do gráfico Número do chamados Por Ano/Fila

```python
def agrupar_filas(fila):
    if fila.startswith('Desenvolvimento'):
        return 'N3'
    elif fila.startswith('Orçamentação'):
        return 'Orçamentação'
    elif fila.startswith('Projeto'):
        return 'Projeto'
    else:
        return fila

df['Fila Agrupada'] = df['Fila'].apply(agrupar_filas)
```

```python
df_agrupado = df.groupby(['Ano', 'Fila Agrupada']).size().reset_index(name='Número de Chamados')
```

```python
import matplotlib.pyplot as plt
import numpy as np

anos = df_agrupado['Ano'].unique()
filas_agrupadas = df_agrupado['Fila Agrupada'].unique()

matriz_dados = []
for fila in filas_agrupadas:
    valores = []
    for ano in anos:
        valor = df_agrupado[(df_agrupado['Ano'] == ano) & (df_agrupado['Fila Agrupada'] == fila)]['Número de Chamados']
        valores.append(valor.values[0] if not valor.empty else 0)
    matriz_dados.append(valores)

largura_barra = 0.15
posicoes = np.arange(len(anos))

plt.figure(figsize=(10, 6))
for i, fila in enumerate(filas_agrupadas):
    plt.bar(posicoes + i * largura_barra, matriz_dados[i], width=largura_barra, label=fila)

plt.xticks(posicoes + largura_barra * (len(filas_agrupadas) - 1) / 2, anos)
plt.xlabel('Ano')
plt.ylabel('Número de Chamados')
plt.title('Número de Chamados por Ano e por Fila Agrupada')
plt.legend(title='Fila Agrupada')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Verificar se colunas existem
if 'Fechamento' in df.columns and 'Abertura' in df.columns:

    # 2. Garantir que datas estão no formato datetime
    df['Abertura'] = pd.to_datetime(df['Abertura'], errors='coerce')
    df['Fechamento'] = pd.to_datetime(df['Fechamento'], errors='coerce')

    # 3. Verificar se 'Status' existe e contém "Fechado"
    if 'Status' in df.columns:
        print("Valores únicos de Status:", df['Status'].unique())
        df_closed = df[df['Status'].str.lower() == 'fechado'].copy()
    else:
        print("⚠️ Coluna 'Status' não encontrada. Usando todas as linhas com datas válidas.")
        df_closed = df.dropna(subset=['Fechamento', 'Abertura']).copy()

    # 4. Calcular duração em horas
    df_closed['Duracao_Hours'] = (df_closed['Fechamento'] - df_closed['Abertura']).dt.total_seconds() / 3600

    # 5. Verificar se coluna 'N123' existe
    if 'N123' in df_closed.columns:
        # Agrupar por 'N123' e somar horas
        horas_por_n123 = df_closed.groupby('N123')['Duracao_Hours'].sum().sort_values(ascending=False)

        print("\n--- Total de Horas Aplicadas por Nível (N123) para Chamados Fechados ---")
        print(horas_por_n123.round(2))

        # 6. Plotar gráfico se houver dados
        if not horas_por_n123.empty:
            plt.figure(figsize=(10, 6))
            sns.barplot(x=horas_por_n123.index, y=horas_por_n123.values, palette='viridis')
            plt.title('Total de Horas Aplicadas por Nível (N123)')
            plt.xlabel('Nível (N123)')
            plt.ylabel('Total de Horas Aplicadas')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
        else:
            print("⚠️ Nenhum dado de duração para chamados fechados com classificação 'N123'.")
    else:
        print("❌ A coluna 'N123' não existe no DataFrame.")
else:
    print("❌ As colunas 'Abertura' e/ou 'Fechamento' não estão presentes no DataFrame.")
```

```python
df['SLA_atendido'] = df['SLA'].map({'sim': True, 'nao': False})
```

```python
# Convert 'Abertura' column to datetime objects
df['Abertura'] = pd.to_datetime(df['Abertura'], errors='coerce')

# Create 'Ano' column
df['Ano'] = df['Abertura'].dt.year

resumo = (
    df.groupby('Ano')
      .agg(
          total_chamados = ('SLA_atendido', 'count'),
          # Fill potential NaN values with False before applying ~ and sum
          fora_SLA = ('SLA_atendido', lambda x: (~x.fillna(False)).sum())
      )
      .reset_index()
)
resumo['percent_fora_SLA'] = 100 * resumo['fora_SLA'] / resumo['total_chamados']
print(resumo)
```

```python
plt.figure(figsize=(8,5))
plt.bar(resumo['Ano'], resumo['fora_SLA'], color='tomato')
plt.title('Chamados Fora do SLA por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Chamados Fora do SLA')
plt.show()
```
